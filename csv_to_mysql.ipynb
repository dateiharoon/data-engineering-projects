```python
import configparser
import pyodbc 
import mysql.connector
#from config import db
import pandas as pd
%load_ext sql
```

```python
# connection to database
config = configparser.ConfigParser()
config.read(r'config.txt')

host  = config.get('DEFAULT','host')
database = config.get('DEFAULT','database')
username  = config.get('DEFAULT','username')
password  = config.get('DEFAULT','password')

db = mysql.connector.connect(
    host = host,
    database = database,
    user = username,
    password = password 
)

cur = db.cursor()
```

```python
## read csv file and convert to dataframe 
df = pd.read_csv('segmentation.csv')
```

```python
# insert df into mysql database 
for i, row in df.iterrows():
    cur.execute(
        '''INSERT INTO segmentation (segment, country, product, discount_band, 
        units_sold, manufacturing_price, sale_price, gross_sales, discounts, sales, 
        cogs, profit, date, month_number, month_name, year) 
         VALUES (%s, %s,%s, %s,%s, %s,%s, %s,%s, %s,%s, %s,%s, %s,%s, %s)''',
        row.tolist())
```

```python

#get total number of row in the db 
    
select_segmentation_data_count = '''SELECT COUNT(*) FROM segmentation'''
cur.execute(select_segmentation_data_count)
records = cur.fetchall()
print(records)
```

```python
#select all rows in segementation table
select_segmentation_data = '''SELECT * FROM segmentation'''
cur.execute(select_segmentation_data)
all_records = cur.fetchall()
```

```python
select_segmentation_data_count = '''select Segment,COUNT(*) from segmentation GROUP BY Segment ORDER BY COUNT(*) DESC;'''
cur.execute(select_segmentation_data_count)
all_records = cur.fetchall()
```
